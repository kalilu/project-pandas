# Clean all dates(date,case_number,case_number_1,case_number_2) 

## Aplicamos una función para encontrar las fechas en estos campos y pasar la columna a fecha

```python
def clean_date(date):
    date = re.search('\d{4,}[-./|]\d{2,}[-./|]\d{2,}', date)
    if date:
        str_date =date.group()
        str_date = str_date[0:4] +'/'+ str_date[5:7] +'/'+ str_date[8:10]
        try: 
            clean_date = datetime.strptime(str_date, '%Y/%m/%d').date()
        except:
            clean_date = None
            pass
    else: 
        clean_date = None
    return(clean_date)
    
df['case_number_c'] = df.case_number.apply(lambda x:clean_date(x))
df['case_number_1_c'] = df.case_number_1.apply(lambda x:clean_date(x))
df['case_number_2_c'] = df.case_number_2.apply(lambda x:clean_date(x))
```

# Clean sex: Aplicamos una función para limpiar los valores inválidos (NA=No aplica)

```python
def clean_sex(sex):
    sex=sex.strip()
    if((sex=='M') | (sex=='F')):
        clean_sex=sex
    elif(sex=='nan' or not sex):
        clean_sex = None
    else:
        clean_sex ='NA'
    return(clean_sex)

df['sex_c']= df.sex.apply(lambda x:clean_sex(str(x)));
```

## Aplicamos dos funciones, una para verificar que la edad es un número y otra para convertir el campo a integer

```python
def is_number(age):
    try:
        age=age.strip()
        int(age)
        return True
    except ValueError:
        return False

def clean_age(age):
    if is_number(age)==True:
        return(int(age))
    else:
        return None

df['age_is_valid_value'] = df['age'].fillna(-1)
df['age_is_valid_value']= df.age_is_valid_value.apply(lambda x:is_number(str(x)))

df['age_c']= df.age.apply(lambda x:clean_age(str(x)))
```

## Categorizamos la variable activity, sin embargo no eliminamos en este caso la descripción 

```python
def clean_activity(activity):
    new_activity = re.sub(r'[^\w\s]',' ',activity)
    clean_activity= [str.lower(w) for w in activity.split() if w.endswith('ing')]
    return(clean_activity)
    
df['activity_c'] = df.activity.apply(lambda x:clean_activity(str(x)))
```

# Clean country

## Haremos la primera búsqueda con la librería country_converter


```python
cc = coco.CountryConverter()

df['country_c']=df.country.apply(lambda x: cc.convert(names = x, to = 'name_short'))
```

## Haremos una segunda búsqueda con la librería pycountry, que tiene todas los paises en una lista

```python
list(pycountry.countries)
```

## Search-fuzzy busca posibles match con el nombre del país y devuelve una lista

```python
def get_country_code(name):
    try:
        country=pycountry.countries.search_fuzzy(name)
    except LookupError:
        return list()
    return [x.name for x in country]

df['country_c_2']=df.country.apply(lambda x: list(get_country_code(str(x))))
```
